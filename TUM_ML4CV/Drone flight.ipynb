```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline 
```

```python
D = np.array([[2, 0, 1],[1.08,1.68,2.38],[-0.83,1.82,2.49],[-1.97,0.28,2.15],[-1.31,-1.51,2.59],[0.57,-1.91,4.32]])
print(D)
```

```python
T = np.arange(6)
X = D[:,0]
Y = D[:,1]
Z = D[:,2]
```

```python
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
```

### Let's assume input data (X) not as T, but as cos(T), and input data (Y) as sin(T). If w_x = w_y, then this coordinates will return constant value for each period of time.

```python
Co = np.append([np.ones(6)],[np.cos(T)],axis=0)
w0 = np.dot(Co,Co.T)
w0 = np.linalg.inv(w0)
w0 = np.dot(np.dot(Co,X),w0)
w_x = w0
def X_func(t):
    return w_x[0]+w_x[1]*np.cos(t)

plt.plot(X)
t1 = np.arange(0.0, 5.0, 0.1)
plt.plot(t1, X_func(t1), 'bo')
```

```python
Co = np.append([np.ones(6)],[np.sin(T)],axis=0)
w0 = np.dot(Co,Co.T)
w0 = np.linalg.inv(w0)
w0 = np.dot(np.dot(Co,Y),w0)
w_y = w0
def Y_func(t):
    return w_y[0]+w_y[1]*np.sin(t)
plt.plot(Y)
t1 = np.arange(0.0, 5.0, 0.1)
plt.plot(t1, Y_func(t1), 'bo')
```

### If speed is const, then traectory must be linear function given T.

```python
Co = np.append([np.ones(6)],[T],axis=0)
w0 = np.dot(Co,Co.T)
w0 = np.linalg.inv(w0)
w0 = np.dot(np.dot(Co,Z),w0)
w_z = w0
def Z_func(t):
    return w_z[0]+w_z[1]*t
plt.plot(Z)
t1 = np.arange(0.0, 5.0, 0.1)
plt.plot(t1, Z_func(t1), 'bo')
```

### If acceleration is const, then traectory must be quadratic function given T. (But z^3 gives better results, btw)

```python
Co = np.append([np.ones(6)],[T],axis=0)
Co = np.append(Co, [np.square(T)],axis=0)
#Co = np.append(Co, [np.square(T)*T],axis=0)
Co = Co.T
w0 = np.dot(Co.T,Co)
w0 = np.linalg.inv(w0)
w0 = np.dot(np.dot(w0,Co.T),Z)
w_z2 = w0
def Z2_func(t):
    return w_z2[0]+w_z2[1]*t+w_z2[2]*t*t
#def Z2_func(t):
    #return w_z2[0]+w_z2[1]*t+w_z2[2]*t*t+w_z2[3]*t*t*t
plt.plot(Z)
t1 = np.arange(0.0, 5.0, 0.1)
plt.plot(t1, Z2_func(t1), 'r')
print(w0)
```

```python
def V_X(t):
    return np.gradient(X_func(t))
def V_Y(t):
    return np.gradient(Y_func(t))
def V_Z(t):
    return np.gradient(Z_func(t))
def V_Z2(t):
    return np.gradient(Z2_func(t))
```

```python
def V(t):
    return np.sqrt(np.square(V_X(t))+np.square(V_Y(t))+np.square(V_Z(t)))
t1 = np.arange(0.0, 5.0, 0.1)
print(V(np.arange(5)))
plt.plot(t1, V(t1), 'bo')
```

### As we can see above, speed is constant with a little epsilon<0.0001. So speed is 1.75

```python
def V2(t):
    return np.sqrt(np.square(V_X(t))+np.square(V_Y(t))+np.square(V_Z2(t)))
def A2(t):
    return np.sqrt(np.square(np.gradient(V_X(t)))+np.square(np.gradient(V_Y(t)))+np.square(np.gradient(V_Z2(t))))
t1 = np.arange(0.0, 5.0, 0.1)
print(A2(np.arange(10)))
plt.plot(t1, A2(t1), 'bo')
```

### As we can see above, acceleration is constant with a little epsilon<0.001.

### Here is a traectory of const speed (blue) and const acceleration (red).

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
t1 = np.arange(0.0, 5.0, 0.1)
ax.plot(X_func(t1), Y_func(t1), Z_func(t1))
ax.plot(X_func(t1), Y_func(t1), Z2_func(t1),'r')
ax.plot(X, Y, Z)
```
